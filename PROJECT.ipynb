# READ ALL FILES IN FOLDER AND APPEND THEM IN 1 FILE

```python
import os
folder=r"C:\Users\User\Documents\file"
Path=[os.path.join(folder,name)for name in os.listdir(folder)]
for w in Path:
    with open(w,'r',encoding="ISO-8859-1")as f:
        files=f.readlines()
newfile=''.join(files)
newfile
```

```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
from nltk.tokenize import word_tokenize
tfile=word_tokenize(newfile)
tfile
```

# REMOVE STOPWORDS

```python
from gensim.parsing.preprocessing import remove_stopwords
rs = remove_stopwords(newfile)
```

```python
rs
```

# LEMMATIZATION

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
l=[]
for w in rs:
    l.append(lemmatizer.lemmatize(w))
```

```python
mod_string = ""
for w in l:
    if w.isalnum() or w == ' ':
        mod_string += w
l = mod_string
print(l)
```

```python
li = l.split(" ")
li
```

# CLUSTERING


```python

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score

v = TfidfVectorizer(stop_words='english')
X = v.fit_transform(li)

true_k = 10
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=1000, n_init=1)
model.fit(X)

order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = v.get_feature_names()
for i in range(true_k):
    print("Cluster %d:" % i),
    for ind in order_centroids[i, :10]:
        print(' %s' % terms[ind]),
```

```python
print("\n")
print("Prediction")
Y = v.transform(["everybody"])
prediction = model.predict(Y)
print(prediction)
```
